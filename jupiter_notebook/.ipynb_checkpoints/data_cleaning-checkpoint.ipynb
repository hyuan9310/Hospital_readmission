```python
import numpy as np
import pandas as pd
import math 
from sklearn.model_selection import train_test_split 
from sklearn.utils import resample
from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing

#visualisation
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv("../../diabetic_data.csv")
raw.head(5)
```

```python
### check missing value
### since the missing value here is '?', rather than None, we do not use print(df.isnull().sum())
print(f"Total rows: {len(raw)}")
for col in raw.columns: 
    print(f"column {col} has {len(raw[raw[col]=='?'])+len(raw[raw[col]=='None'])} missing value,which occupies {(len(raw[raw[col]=='?'])+len(raw[raw[col]=='None']))/len(raw)*100}%")
```

```python
### drop weight,medical_specialty,payer_code columns
### manually drop some meaning less columns
### drop ['encounter_id','payer_code','weight','patient_nbr','diag_3','medical_specialty']
df = raw.drop(['encounter_id','patient_nbr','payer_code','weight','diag_2','diag_3',
               'max_glu_serum','A1Cresult','medical_specialty'], axis=1, inplace=False)

## drop rows with '?'
a = []
for col in df.columns:
    a = a + df.index[df[col] == '?'].tolist()
    if len(df.index[df[col] == '?'].tolist())>0:
        print(f"drop missing value columns {col}")
print(f"total dropping row is {len(a)}")

b = df.index[df['gender'] == 'Unknown/Invalid'].tolist()
removeRow = a + b
df.drop(removeRow, axis = 0,inplace=True)
```

```python
for col in df.columns:
    print(f'{col}:{df[col].value_counts()}')
```

```python
### mannually drop columns based on number of the majority  for each feature
### second majority counts<1100
drop_cols_mannual = ['admission_source_id','repaglinide','nateglinide','chlorpropamide','acetohexamide','tolbutamide','acarbose','miglitol',
                    'troglitazone','tolazamide','examide','citoglipton','glyburide-metformin','glipizide-metformin','glimepiride-pioglitazone',
                     'metformin-rosiglitazone','metformin-pioglitazone']
df.drop(drop_cols_mannual,axis =1, inplace =True)

# concatenate extreme small unique features into 'other' for each columns
df['admission_type_id'] = df['admission_type_id'].apply(lambda x: 'other' if x in [8,7,4] else x)
df['discharge_disposition_id'] = df['discharge_disposition_id'].apply(lambda x: 'other' if x in [5,25,4,7,23,13,14,28,8,15,24,9,17,16,19,10,27,12,20] else x)

df.head(5)
```

```python
### manually select categorical data and numerical data
categorical_data_cols = ['race','gender','age','admission_type_id', 'discharge_disposition_id','metformin', 'glimepiride', 'glipizide', 'glyburide', 'pioglitazone', 'rosiglitazone', 'insulin','change', 'diabetesMed']
numerical_data_cols = [ 'time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications', 'number_outpatient', 'number_emergency', 'number_inpatient', 'diag_1', 'number_diagnoses']
label_col = ['readmitted']
```

### categorical_data_cols

```python
### check categorical data column unique values:
categorical_dict = {}
for col in categorical_data_cols:
    categorical_dict[col] = len(df[col].unique())
    print(f"column {col}: {df[col].unique()}")
print(f'\ncategorical_dict: \n {categorical_dict}')
```

```python
### data transform
### transform label to binary class:{ 1: readmitted, 2: not readmitted}
df_transform = df.copy()
df_transform['readmitted'] = np.where(df_transform['readmitted'] == '<30',1,0)

### race, get dummy 
one_hot_4 = pd.get_dummies(df_transform['race'])
df_transform = df_transform.join(one_hot_4)
df_transform = df_transform.drop(['race'],axis = 1)
# try:
#     del categorical_dict['race']
# except KeyError:
#     pass

### age: use average age:
df_transform['age'].replace('[0-10)',5,inplace=True)
df_transform['age'].replace('[10-20)',15,inplace=True)
df_transform['age'].replace('[20-30)',25,inplace=True)
df_transform['age'].replace('[30-40)',35,inplace=True)
df_transform['age'].replace('[40-50)',45,inplace=True)
df_transform['age'].replace('[50-60)',55,inplace=True)
df_transform['age'].replace('[60-70)',65,inplace=True)
df_transform['age'].replace('[70-80)',75,inplace=True)
df_transform['age'].replace('[80-90)',85,inplace=True)
df_transform['age'].replace('[90-100)',95,inplace=True)
# try:
#     del categorical_dict['age']
# except KeyError:
#     pass

###gender
df_transform['gender'] = np.where(df_transform['gender']=='Male',1,0)
# del categorical_dict['gender']

### admission_type_id, get dummy and drop largest one to avoid multicollinearity
one_hot_5 = pd.get_dummies(df_transform['admission_type_id'])
one_hot_5.columns = [f"admission_type_{i}" for i in [1,2,3,5,6,'other']]
one_hot_5 = one_hot_5.drop(columns = ['admission_type_1'])
df_transform = df_transform.join(one_hot_5)
df_transform = df_transform.drop(['admission_type_id'],axis = 1)
# try:
#     del categorical_dict['admission_type_id']
# except KeyError:
#     pass

### discharge_disposition_id, get dummy 
one_hot_6 = pd.get_dummies(df_transform['discharge_disposition_id'])
one_hot_6.columns = [f"discharge_disposition_{i}" for i in [1,2,3,6,11,18,22,'other']]
one_hot_6 = one_hot_6.drop(columns = ['discharge_disposition_1'])
df_transform = df_transform.join(one_hot_6)
df_transform = df_transform.drop(['discharge_disposition_id'],axis = 1)

df_transform.head(5)
```

```python
### ['No' 'Steady' 'Up' 'Down'] = [0,2,3,1]
### transform categorical to numerical
def medicine_volumn(series):
    series = series.apply(lambda x: 0 if x=='No' else(2 if x=='Steady' else (3 if x=='Up' else 1)))
    return series

for col in ['metformin', 'glimepiride', 'glipizide', 'glyburide', 'pioglitazone', 'rosiglitazone', 'insulin',  'glimepiride-pioglitazone']:
    try:
        df_transform[col] = medicine_volumn(df_transform[col])
    except:
        print(col)
        
df_transform.head(5)
```

### numerical data

```python
#diag_1: transform to different type of illness
diag_1 = df_transform['diag_1']
#Circulatory 390â€“459, 785
C = list(range(390,460)) + [785]
C = [str(x) for x in C]

for i in C:
    diag_1.replace(i,'Circulatory',inplace = True)

#Respiratory	460â€“519, 786
R = list(range(460,520))+[786]
R = [str(i) for i in R ]

for i in R:
    diag_1.replace(i,'Respiratory', inplace = True)

#Digestive	520â€“579, 787
D = list(range(520,580))+[787]
D = [str(i) for i in D]
    
for i in D:
    diag_1.replace(i,'Digestive',inplace = True)
    
#Diabetes	250.xx
DB = list(np.arange(250.01,251,0.01))
DB = [round(i,2) for i in DB]
DB = [str(i) for i in DB]
DB += ['250']

for i in DB:
    diag_1.replace(i,'Diabetes',inplace = True)
    
#Injury	800â€“999
I = list(range(800,1000))
I = [str(i) for i in I]

for i in I:
    diag_1.replace(i,'Injury',inplace = True)
    
#Musculoskeletal	710â€“739
M = list(range(710,740))
M = [str(i) for i in M]

for i in M:
    diag_1.replace(i,'Musculoskeletal',inplace = True)

#Genitourinary	580â€“629, 788
G = list(range(580,630))+['788']
G = [str(i) for i in G]

for i in G:
    diag_1.replace(i,'Genitourinary',inplace = True)
#Neoplasms 140â€“239
N = list(range(140,240))
N = [str(i) for i in N]

for i in N:
    diag_1.replace(i, 'Neoplasms', inplace= True)

diagSet = set(['Circulatory','Respiratory','Digestive','Diabetes','Injury','Musculoskeletal','Genitourinary','Neoplasms'])
oindex = df_transform['diag_1'].loc[~df_transform['diag_1'].isin(diagSet)].index.tolist()
for i in oindex:
    df_transform.at[i,'diag_1'] = 'other_diag'

one_hot_6 = pd.get_dummies(df_transform['diag_1'])
df_transform = df_transform.join(one_hot_6)
df_transform = df_transform.drop('diag_1',axis = 1)
# try:
#     numerical_data_cols.remove('dial_1')
# except ValueError:
#     pass
```

### normalization

```python
### cols need norm
norm_cols = ['age','time_in_hospital', 'num_lab_procedures', 'num_procedures',
       'num_medications', 'number_outpatient', 'number_emergency',
       'number_inpatient', 'number_diagnoses', 'metformin', 'glimepiride',
       'glipizide', 'glyburide', 'pioglitazone',
       'rosiglitazone', 'insulin']
dict_norm_max_min = {}

for col in norm_cols:
    max_ ,min_ = df_transform[col].max(),df_transform[col].min()
    dict_norm_max_min[col] = (max_, min_)
    df_transform[col] = (df_transform[col]-min_)/(max_ - min_)

# store max min info as csv file
max_min_info = pd.DataFrame(dict_norm_max_min,index=['max', 'min'])
max_min_info.to_csv('../data/max_min_norm.csv')
```

### imbalanced data

```python
### check data balance, if not, resample data to balanced data
### class 1: readmitted, class 0: not readmitted

def resampletrain(df):
    print(len(df[df.readmitted==0]) , len(df[df.readmitted==1]))
    if len(df[df.readmitted==0]) >= len(df[df.readmitted==1]):
        ### majority is class 0
        df_majority = df[df.readmitted==0]
        df_minority = df[df.readmitted==1]
    else:
        ### majority is class 1
        df_majority = df[df.readmitted==1]
        df_minority = df[df.readmitted==0]
        
    df_majority_downsampled = df_majority.sample(replace=False, n=len(df_minority),random_state = 123)
    df_downsampled = pd.concat([df_majority_downsampled, df_minority])
    return df_downsampled

print(f"before sample data: \n{df_transform['readmitted'].value_counts()}")
df_transform = resampletrain(df_transform)
print(f"\nafter sample data: \n{df_transform['readmitted'].value_counts()}")
```

### reset index and save to csv

```python
# cleaned data frame
df_transform = df_transform.reset_index(drop=True)
df_transform.to_csv('../data/cleaning_data.csv')
df_transform[['readmitted']].to_csv('../data/cleaning_data_label.csv')
df_transform.loc[:,df_transform.columns != 'readmitted'].to_csv('../data/cleaning_data_features.csv')
```
